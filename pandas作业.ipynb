```python
import pandas as pd

# 基础DataFrame
df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})
group_agg = df.groupby("gender")["score"].agg(["mean", "max"])
print(group_agg)
```

```python
group_count = df.groupby("gender").size()
# 也可以用 df["gender"].value_counts()
print(group_count)
```

```python
df_sorted = df.sort_values(by="score", ascending=True)
print(df_sorted)
```

```python
# 添加重复行（Tom, 85, 男）
new_row = pd.DataFrame([["Tom", 85, "男"]], columns=df.columns)
df_with_dup = pd.concat([df, new_row], ignore_index=True)
print("添加重复行后：")
print(df_with_dup)

# 去重
df_dedup = df_with_dup.drop_duplicates()
print("\n去重后：")
print(df_dedup)
```

```python
import pandas as pd

# 基础数据
df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})

# 1. 创建时间序列并转为datetime格式
time_list = ["2025-01-01", "2025-01-02", "2025-01-03"]
df['date'] = pd.to_datetime(time_list)

print("转换后的时间列：")
print(df['date'])
```

```python
# 2. 提取时间列的月份、星期（同时展示数字和中文名称）
df['month'] = df['date'].dt.month
# dt.dayofweek：0=周一，1=周二，2=周三，3=周四，4=周五，5=周六，6=周日
df['weekday_num'] = df['date'].dt.dayofweek
# dt.day_name()：获取英文星期名；map映射为中文
df['weekday_cn'] = df['date'].dt.day_name().map({
    'Monday': '周一',
    'Tuesday': '周二',
    'Wednesday': '周三',
    'Thursday': '周四',
    'Friday': '周五',
    'Saturday': '周六',
    'Sunday': '周日'
})

print("\n提取月份和星期后：")
print(df[['date', 'month', 'weekday_num', 'weekday_cn']])
```

```python
# 计算均值和标准差，完成标准化
score_mean = df['score'].mean()
score_std = df['score'].std()
df['score_standardized'] = (df['score'] - score_mean) / score_std

print("标准化后的score列：")
print(df[['score', 'score_standardized']])
```

```python
# 创建透视表
pivot_result = pd.pivot_table(
    df,
    index='gender',   # 行索引
    values='score',   # 值列
    aggfunc='median'  # 聚合函数：中位数
)

print("\n透视表结果：")
print(pivot_result)
```
